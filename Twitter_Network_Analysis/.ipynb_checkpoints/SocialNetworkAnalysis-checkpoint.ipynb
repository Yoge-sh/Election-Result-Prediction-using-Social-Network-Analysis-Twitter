# NETWORK ANALYSIS

```python
d1['SupEmo']=d1['Emotion'].copy()

for i in range(0,25688):
  if d1['Emotion'].iloc[i]=='pos':
    d1['SupEmo'].iloc[i]='nmp'
  elif d1['Emotion'].iloc[i]=='neg':
    d1['SupEmo'].iloc[i]='nmn'
```

```python
d2['SupEmo']=d2['Emotion'].copy()

for i in range(0,14261):
  if d2['Emotion'].iloc[i]=='pos':
    d2['SupEmo'].iloc[i]='rgp'
  elif d2['Emotion'].iloc[i]=='neg':
    d2['SupEmo'].iloc[i]='rgn'
```
