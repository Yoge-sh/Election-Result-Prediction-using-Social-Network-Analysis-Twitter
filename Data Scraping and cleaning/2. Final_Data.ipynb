```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Common keywords along with hashtags
df1=pd.read_csv("/content/drive/MyDrive/Graph_theory_and_social_networks/TweetsElections2019_2.csv",index_col=0)

#TWEETS WITH POPULAR HASHTAGS
df2=pd.read_csv('/content/drive/MyDrive/Graph_theory_and_social_networks/TweetsElections2019Data2_2.csv',index_col=0)

#LokSabhaElections2019
df3=pd.read_csv('/content/drive/MyDrive/Graph_theory_and_social_networks/Tweets#LokSabhaElections2019_2.csv',index_col=0)
```

```python
df = pd.concat([df1,df2,df3],axis=0)
df=df.reset_index(drop=True)
df.head()
```

```python
df
```

```python
df.info()
```

# Data Cleaning

```python
df = df.drop_duplicates(keep='first',subset=['Tweet'])
df=df.reset_index(drop=True)
df.head()
```

```python
df.info()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
Word=['https','co']
text = " ".join(tweet for tweet in df["Tweet"])
word_cloud = WordCloud(collocations = False, background_color = 'black', colormap = 'Wistia', min_font_size = 8).generate(text)
plt.figure(figsize = (10, 8))
plt.imshow(word_cloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

```python
word_cloud.to_file("wcloud.png",dpi=100)
```
